# Generating Visualizations
This is not the main file. This is just a file where we generate the visualizations for the paper. Please refer to `main.py` to train and/or test the model.

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
def plot_history_graph(
    csv_filename: str,
    title: str,
    data_dir: str = './histories/'
):
    path: str = os.path.join(data_dir, csv_filename)
    df: pd.DataFrame = pd.read_csv(path)
    plt.plot(df.index, df['Train Loss'], label='Train')
    plt.plot(df.index, df['Val Loss'], label='Validation')
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.title(title)
    plt.ylim(0.75, 2.2)
    plt.legend()
    plt.show()
```

```python
plot_history_graph("history_dropout.csv", "Blockwise Dropout Curves")
```

```python
plot_history_graph("history_dropout_global.csv", "Global Dropout Curves")
```

```python
plot_history_graph("history_adam.csv", "Adam Loss Curves")
plot_history_graph("history_adamw.csv", "AdamW Loss Curves")
plot_history_graph("history_sgd.csv", "SGD Loss Curves")
```
